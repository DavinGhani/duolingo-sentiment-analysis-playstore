```python
import pandas as pd
from google_play_scraper import reviews_all, Sort
```

```python
# --- PENGATURAN ---
app_id_target = 'com.duolingo'
jumlah_maksimal_ulasan = 15000

print(f"Memulai scraping ulasan untuk aplikasi: {app_id_target}")
```

```python
# --- PROSES SCRAPING ---
result = reviews_all(
        app_id_target,
        lang='id',
        country='id',
        sort=Sort.NEWEST,
        sleep_milliseconds=0
    )
df_reviews = pd.DataFrame(result)
print(f"\n✅ Scraping selesai. Total ulasan yang ditemukan: {len(df_reviews)}")

# --- BATASI JUMLAH DATA ---
print(f"Membatasi data menjadi {jumlah_maksimal_ulasan} ulasan terbaru...")
df_limited = df_reviews.head(jumlah_maksimal_ulasan)

# --- PROSES AKHIR ---
# Pilih dan ganti nama kolom yang relevan
df_limited = df_limited[['userName', 'score', 'at', 'content']]
df_limited = df_limited.rename(columns={'score': 'rating', 'at': 'date', 'content': 'comment'})

print(f"✅ Data berhasil dibatasi. Jumlah akhir: {len(df_limited)}")
```

```python
# --- HASIL ---
print("\nContoh 5 data pertama:")
print(df_limited.head())
```

```python
# Simpan hasil ke file CSV
csv_filename = f'reviews_playstore_{app_id_target}_{len(df_limited)}.csv'
df_limited.to_csv(csv_filename, index=False)
print(f"\n✅ Data berhasil disimpan ke file: {csv_filename}")
```
